Recentemente, a associação recebeu diversas reclamações de inquilinos de uma determinada área da cidade. Eles alegam que suas taxas mensais de aluguel estão muito mais altas se comparadas com as de outras partes da cidade.

Sua equipe deve investigar a veracidade das alegações. Para isso, reuniu os dados da tabela a seguir.

A área A representa a área da cidade onde os inquilinos estão insatisfeitos
com seus aluguéis mensais. Os dados representam os aluguéis mensais pagos por uma amostra aleatória de inquilinos da área A e as três outras áreas de tamanho similar. Considere que todos os apartamentos representados são, aproximadamente, do mesmo tamanho e com as mesmas comodidades.

Outra possível fonte de informações para embasamento de possíveis justificativas

Atividade Estatística Descritiva 02.png

```python
'''Criação do conjunto de dados'''
import pandas as pd
# Listas com os aluguéis de cada área
areaA = [1275, 1110, 975, 862, 1040, 997, 1119, 908, 890, 1055, 860, 975]
areaB = [1124, 954, 815, 1078, 843, 745, 796, 816, 938, 1082, 750, 703]
areaC = [1085, 827, 793, 1170, 919, 943, 756, 765, 809, 1020, 710, 775]
areaD = [928, 1096, 862, 735, 798, 812, 1232, 1036, 998, 914, 1005, 930]
# Concatenando as listas em um Pandas DataFrame
aluguel = pd.DataFrame({'Área A': areaA, 'Área B': areaB,
                        'Área C': areaC, 'Área D': areaD})
```

# Distrubuição dos aluguéis por cada área
Construa histogramas representativos das distribuições de aluguéis por cada área da cidade, usando:

*   Extremo inferior da primeira classe: US\$ 700,00
*   Extremo superior da última classe: US\$ 1.300,00
*   Amplitude classe: US\$ 100,00
*   Escala do eixo horizontal: pontos médios de cada classe
*   Escala do eixo vertical: 0 a 5

```python
'''Construção dos histogramas das distribuições de aluguel'''
import matplotlib.pyplot as plt
# Constantes para construção das classes
MIN, MAX, AMP = 700, 1300, 100
# Determinação das classes
classes = pd.interval_range(start=MIN, end=MAX, freq=AMP)
# Pontos médios das classes
pts_medios = [classe.mid for classe in classes]
# Histogramas (um diagrama para cada área registrada no painel aluguel)
histograma = aluguel.hist(bins=[classe.left for classe in classes]+
                          [classes[-1].right],
                          color='blue', edgecolor='black',
                          grid=False, figsize=(8,8))
# Atributos para cada histograma (indexados como uma matriz iniciando em (0,0))
indices = [(i,j) for i in range(2) for j in range(2)]
for (i,j) in indices:
  histograma[i][j].set(xticks=pts_medios, yticks=list(range(6)),
                       xlabel='Aluguel mensal [US$]', ylabel='Frequência')
''' IMPORTANTE: Não era necessário exibir todos os histogramas em uma mesma
figura, como mostrado aqui. '''
```

# Medidas centrais e de variação das distribuições
Calcule as medidas centrais e de variação das distribuições de aluguéis em cada área da cidade.

```python
'''Podemos usar diversas sintaxes para obter as medidas centrais e
de distribuição. Aqui, optamos por duas formas simples.'''
# Uso do método describe() para o painel aluguel
estats = aluguel.describe()
display(estats)
# Como uma alternativa, é possível usar o método agg()
aluguel.agg(['mean', 'median', 'std'])
```

# Quartis e Boxplots
Determine os quartis para a distribuição de aluguéis de cada cidade. A seguir, construa os respectivos diagramas boxplot.

```python
'''O método quantile() do painel aluguel permite o cálculo dos percentis
desejados. No caso, queremos os quartis das distribuições. Então devemos
passar uma lista com 25, 50 e 75%.'''
# Quartis 
display(aluguel.quantile([0.25, 0.5, 0.75]))
# Construção dos boxplots
box = aluguel.boxplot(showmeans=True)
plt.title('Distribuição dos aluguéis por área')
plt.ylabel('Aluguel [US$]')
```
